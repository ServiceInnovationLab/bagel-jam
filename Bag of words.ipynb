```python
from wordcounter.wordcounter import WordCounter 
import re
import nltk
nltk.download('wordnet')
nltk.download('stopwords')

from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import textcleaner as tc

# files are in ./bag-of-words.txt
words = {}
bag_of_words_file = 'bag-of-words.txt'
```

```python
# with open(bag_of_words_file) as f:
#     contents = f.read()
#     contents = contents.replace('\n', ' ')
#     contents = re.sub('[^0-9a-zA-Z훮]+', ' ', contents)
#     contents = contents.lower()
#     data = tc.document(contents)
#     contents = data.remove_stpwrds()

contents = tc.main_cleaner(bag_of_words_file, 'words')

print(contents)
```

```python
wnl = WordNetLemmatizer()

tokens = word_tokenize(" ".join(contents))

lemmatized_words = [wnl.lemmatize(token) for token in tokens]

word_counter = WordCounter(" ".join(lemmatized_words))
freq = word_counter.get_word_frequencies()

s = sorted(freq.items(), key=lambda kv: kv[1])

print(s)
```
